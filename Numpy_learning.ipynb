```python
import numpy as np

a = np.array([1,2,3,4,5])

print(a)
```

```python
b = np.array([[1,2,3,4],
              [5,6,7,8]
              ])
print(b)
```

```python
c = np.array([[
    [1,2,3,],
     [4,5,6]
    ],
     [
         [3,4,5],
          [9,7,8]
         ]
              ])
print(c)
print("Diamention:",c.ndim)
```

```python
d = np.array([[1.1, 3.4, 8.4],[1.1, 3.4, 8.4]], dtype=float)
print(d)
```

```python
arng = np.arange(1,10,2)
print(arng)
```

```python
re_shape = np.arange(1,13).reshape(4,3)
print(re_shape)
```

```python
a1 = np.arange(10)
a2 = np.arange(12, dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```
